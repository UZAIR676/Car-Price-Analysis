```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

Data Collection and processing

```python
data = pd.read_csv('/content/car data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# encoding the fuel_type column
print(data.Fuel_Type.value_counts())
# encoding the seller_type column
print(data.Seller_Type.value_counts())
# # encoding the transmission column
print(data.Transmission.value_counts())
```

```python
# now we are converting the letter to numbers
data.replace({'Fuel_Type':{"Petrol":0,"CNG":2,'Diesel':1 }},inplace=True)
data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
data.replace({'Transmission':{'Manual':0,'Automatic':1}})
```

```python
# spliting the training and test data
x = data.drop(['Car_Name','Selling_Price'],axis=1)
y= data['Selling_Price']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# prediction
training_data_prediction = model.predict(x_train)
```

```python
from sre_constants import error
# r squar metrix
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R square ",error_score)
```

visulizing the actual value

```python
plt.scatter(y_train,training_data_prediction)
plt.xlabel('actual price')
plt.ylabel('predicted price')
plt.title('actual proce vs predicted ')
plt.show()
```

```python
test_data_prediction = model.predict(x_test)
```

```python
error_score = metrics.r2_score(y_test, test_data_prediction)  # Use y_test
print("R square:", error_score)
```

```python
plt.scatter(y_test,test_data_prediction)
plt.xlabel('actual price')
plt.ylabel('predicted price')
plt.title('actual proce vs predicted ')
plt.show()
```
